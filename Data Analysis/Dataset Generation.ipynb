```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Stock

```python
data_direc = '../Datasets/TSB-AD-U/'
filename1 = '149_Stock_id_1_Finance_tr_500_1st_7.csv'
#filename2 = '151_Stock_id_3_Finance_tr_500_1st_62.csv'
filename3 = '152_Stock_id_4_Finance_tr_500_1st_2.csv'
#filename4 = '153_Stock_id_5_Finance_tr_500_1st_1.csv'
filename5 = '154_Stock_id_6_Finance_tr_500_1st_38.csv'

df1 = pd.read_csv(data_direc + filename1).dropna()
#df2 = pd.read_csv(data_direc + filename2).dropna()
df3 = pd.read_csv(data_direc + filename3).dropna()
#df4 = pd.read_csv(data_direc + filename4).dropna()
df5 = pd.read_csv(data_direc + filename5).dropna()
```

```python
df = pd.concat([df1,df3,df5])
df.index = range(df.shape[0])
```

```python
df.to_csv('../Datasets/Generated Data/ADG_Stock.csv',index = False)
```

```python
data = df.iloc[:, 0:-1].values.astype(float)
label = df['Label'].astype(int).to_numpy()

dataList = data
labelList = label
labelList = labelList.astype('bool')
```

```python
x = np.array(list(range(dataList.shape[0])))
x1 = x[labelList]
y = dataList
y1 = y[labelList]
fig,ax = plt.subplots(figsize = (24,6))
ax.plot(x,y)
ax.scatter(x1,y1,color='r')
ax.set_xlim(3048,3096)
```
